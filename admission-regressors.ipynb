```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_admission = pd.read_csv('./Admission_Predict_Ver1.1.csv')
data_admission.isnull().sum()
```

```python
data_admission.head()
```

```python
data_admission.drop(['Serial No.'], inplace=True, axis=1)
data_admission.describe()
```

```python
data_admission.columns
```

### Data visualization
We proceed to the visualization of all the variables to see their influence on the chances of admission

```python
sns.set_theme(style='darkgrid')
plt.figure(figsize=(19,12))
plt.subplot(2,3,1)
plt.scatter(data_admission['GRE Score'], data_admission['Chance of Admit '])
plt.title('GRE Score')
plt.subplot(2,3,2)
plt.scatter(data_admission['TOEFL Score'], data_admission['Chance of Admit '])
plt.title('TOEFL Score')
plt.subplot(2,3,3)
plt.scatter(data_admission['University Rating'], data_admission['Chance of Admit '])
plt.title('University Rating')
plt.subplot(2,3,4)
plt.scatter(data_admission['SOP'], data_admission['Chance of Admit '])
plt.title('SOP')
plt.subplot(2,3,5)
plt.scatter(data_admission['LOR '], data_admission['Chance of Admit '])
plt.title('LOR')
plt.subplot(2,3,6)
plt.scatter(data_admission['CGPA'], data_admission['Chance of Admit '])
plt.title('CGPA')
```
